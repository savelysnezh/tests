1) *Рекламная кампания стартовала вчера с дневным бюджетом 40 $. Половина бюджета была израсходована к полудню, <br>
а 80% оставшегося бюджета было потрачено между полуднем и временем закрытия. Сколько долларов не было потрачено?*

```python
# Начальный бюджет
budget = 40

# Вычитаем половину, которую потратили к полудню
budget /= 2

# Вычитаем еще 80% от остатка. Эти деньги мы потратили между полуднем и закрытием
budget -= budget*0.8

# Посмотрим на остаток
f'Остаток: {budget} $'
```

3) *Дизайнеры создали 200 рекламных баннеров для двух рекламных кампаний. 80 из них не использовались <br>
ни в кампании No1, ни в кампании No2, 60 использовались только в кампании No1. И для каждого рекламного <br>
баннера, который использовался в обеих кампаниях, приходится 3 баннера, которые использовались только в <br>
кампании No2. Сколько баннеров было использовано в обеих рекламных кампаниях?*

```python
# Всего баннеров
banners = 200

# Минус те, что не использовались
banners -= 80

# Минус те, что использовались только в кампании №1
banners -= 60

# Далее, по условию, понимаем что на 1 общий баннер было 3 для кампании №2
# Получается, что поделив полученные на прошлом этапе баннеры на 4 "кучи"
# Мы отдаем 3 "кучи" (75%) второй кампании, а остальные будут общими.
banners /= 4

# Вывод результата (общие баннеры)
banners
```

5) *Средний итоговый балл стажера по 4 модулям составляет 78 баллов. Сколько баллов должен получить стажер <br>
за 5-й модуль, чтобы средний балл по всемузаданию составил 80?*

```python
# Мы знаем, что (x1+x2+x3+x4)/4 = 78, где xn - баллы по предмету n
# Избавимся от знаменателя: (x1+x2+x3+x4) = 312
# Обозначим (x1+x2+x3+x4) как "a", тогда
# (a+x5)/5 = 80
# Избавимся от знаменателя и подставим 312 вместо "a". Решим уравнение и получим ответ
# 312+x5=400
# 400-312 = 88
```

был обратный путь? Ответ указать в минутах.*

```python
# Время пути "туда"   : 260/80  = 3.25 часа
# Время пути "обратно": 260/100 = 2.6  часа
# Разница времени в часах: 3.25 - 2.6 = 0.65 часа
# В минутах: 0.65*60 = 39
```
